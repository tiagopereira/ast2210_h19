```python
import numpy as np
import sunpy.map
from sunpy.instr.aia import aiaprep
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
uu = sunpy.map.Map("~/sunpy/data/aia.lev1_uv_24s.2014-02-25T205442Z.3.image_lev1.fits")
```

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [7, 7]
```

```python
tmp = uu.plot()
plt.show()
```

```python
# Make animation
from matplotlib import animation
from IPython.display import HTML

#fig = plt.figure(figsize=(7, 7))
tmp = uu.plot()
ax = plt.gca()

def animate(i):
    uga = uu.data * 2 * i
    ax.set_title("Factor = %i" % (2 * i))
    tmp.set_data(uga)
    return (tmp,)

anim = animation.FuncAnimation(plt.gcf(), animate, init_func=None,
                               frames=20, interval=40, blit=True)
plt.close(anim._fig)
HTML(anim.to_html5_video())
```
